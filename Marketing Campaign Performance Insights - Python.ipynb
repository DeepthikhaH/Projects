```python
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)

avg_roi_by_company = df.groupby('Company')['ROI'].mean().reset_index()


sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_roi_by_company, x='Company', y='ROI', palette='viridis')

plt.title('Average ROI by Company')
plt.xlabel('Company')
plt.ylabel('Average ROI')

plt.xticks(rotation=45)
plt.show()
```

```python
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)

corr = df[['Engagement_Score', 'Conversion_Rate']].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

plt.title('Correlation between Engagement Score and Conversion Rate')
plt.show()
```

```python
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)

avg_conversion = df.groupby(['Language', 'Customer_Segment'])['Conversion_Rate'].mean().reset_index()

idx = avg_conversion.groupby('Language')['Conversion_Rate'].idxmax()
top_segments = avg_conversion.loc[idx]

plt.figure(figsize=(8, 6))
sns.barplot(data=top_segments, x='Language', y='Conversion_Rate', hue='Customer_Segment', dodge=False)

plt.title('Customer Segment with Highest Conversion Rate per Language')
plt.xlabel('Language')
plt.ylabel('Average Conversion Rate')

plt.legend(title='Customer Segment')
plt.show()
```

```python
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

avg_conversion_by_date_company = df.groupby(['Date', 'Company'])['Conversion_Rate'].mean().reset_index()

plt.figure(figsize=(10, 8))
sns.lineplot(data=avg_conversion_by_date_company, x='Date', y='Conversion_Rate', hue='Company')

plt.title('Average Conversion Rate over Time by Company')
plt.xlabel('Date')
plt.ylabel('Average Conversion Rate')

plt.xticks(rotation=45)
plt.legend(title='Company')
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

avg_engagement_by_date = df.groupby('Date')['Engagement_Score'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.lineplot(data=avg_engagement_by_date, x='Date', y='Engagement_Score')

plt.title('Average Engagement Score over Time')
plt.xlabel('Date')
plt.ylabel('Average Engagement Score')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)

avg_acquisition_cost = df.groupby('Location')['Acquisition_Cost'].mean().sort_values(ascending=False).reset_index()

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.barplot(data=avg_acquisition_cost, x='Location', y='Acquisition_Cost', palette='coolwarm')

plt.title('Average Acquisition Cost by Location')
plt.xlabel('Location')
plt.ylabel('Average Acquisition Cost')

plt.xticks(rotation=45)
plt.show()
```
